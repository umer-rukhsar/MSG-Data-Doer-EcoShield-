```python
import numpy as np 
import pandas as pd
```

```python
#Climate Data : Yield
```

```python
df_yield = pd.read_csv('yield.csv', delimiter=';')
df_yield.shape
```

```python
df_yield.head()
```

```python
df_yield.tail()
```

```python
# rename columns.
df_yield = df_yield.rename(index=str, columns={"Value": "hg/ha_yield"})
df_yield.head()
```

```python
# drop unwanted columns.
df_yield = df_yield.drop(['Year Code','Element Code','Element','Year Code','Area Code','Domain Code','Domain','Unit','Item Code'], axis=1)
df_yield.head()
```

```python
df_yield.describe()
```

```python
df_yield.info()
```

```python
#Climate Data : Rainfall
```

```python
df_rain = pd.read_csv('rainfall.csv', delimiter=";")
df_rain.head()
```

```python
#renaming column "Area"  
df_rain = df_rain.rename(index=str, columns={" Area": 'Area'})
```

```python
# drop unwanted columns.
df_rain = df_rain.loc[:, ~df_rain.columns.str.contains('^Unnamed')]
```

```python
# check data types 
df_rain.info()
```

```python
# convert average_rain_fall_mm_per_year from object to float
df_rain['average_rain_fall_mm_per_year'] = pd.to_numeric(df_rain['average_rain_fall_mm_per_year'],errors = 'coerce')
# Convert Year column in yield_df to int64 using pd.to_numeric
df_rain['Year'] = pd.to_numeric(df_rain['Year'], errors='coerce').astype('Int64')
df_rain.info()
```

```python
df_rain = df_rain.dropna()
```

```python
df_rain.describe()
```

```python
# merge yield dataframe with rain dataframe by year and area columns 
yield_df = pd.merge(df_yield, df_rain, on=['Year','Area'])
```

```python
yield_df.shape
```

```python
yield_df.head()
```

```python
yield_df.describe()
```

```python
df_pes = pd.read_csv('pesticides.csv', delimiter=';')
df_pes.head()
```

```python
#renaming column Value
df_pes = df_pes.rename(index=str, columns={"Value": "pesticides_tonnes"})

#droping unnecessary columns
df_pes = df_pes.drop(['Element','Domain','Unit','Item'], axis=1)
df_pes.head()
```

```python
df_pes.describe()
```

```python
df_pes.info()
```

```python
# Convert Year column in yield_df to int64 using pd.to_numeric
df_pes['Year'] = pd.to_numeric(df_pes['Year'], errors='coerce').astype('Int64')
```

```python
# merge Pesticides dataframe with yield dataframe 
yield_df = pd.merge(yield_df, df_pes, on=['Year','Area'])
yield_df.shape
```

```python
yield_df.head()
```

```python
import sklearn
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
correlation_data=yield_df.select_dtypes(include=[np.number]).corr()

mask = np.zeros_like(correlation_data, dtype=bool)
mask[np.triu_indices_from(mask)] = True

f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.palette="vlag"

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(correlation_data, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5});
```
